```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#to ignore warnings
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv('used_cars_data.csv')
# will display the top 5 observations of the dataset
data.head()
```

```python
# Remove S.No. column from data
data = data.drop(['S.No.'], axis = 1)
data.info()
```

```python
# a new column, “Car_Age” to know the age of the car
from datetime import date
date.today().year
data['Car_Age']=date.today().year-data['Year']
data.head()
```

```python
#spliting the name and introduce new variables “Brand” and “Model”
data['Brand'] = data.Name.str.split().str.get(0)
data['Model'] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2)
data[['Name','Brand','Model']]
```

```python
print(data.Brand.unique())  #prints an array of unique values from Brand
print(data.Brand.nunique())  #prints the number of unique values from Brand
```

```python
searchfor = ['Isuzu' ,'ISUZU','Mini','Land']  # Define a list of brand names to search for

# Filter the DataFrame to find rows where the 'Brand' column contains any of the specified keywords
data[data.Brand.str.contains('|'.join(searchfor))].head(5)
```

```python
data["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper","Land":"Land Rover"}, inplace=True)
```

```python
data.describe(include='all').T   #generates descriptive statistics for all columns
```

```python
# Select columns with categorical (object) data types
cat_cols=data.select_dtypes(include=['object']).columns

# Select columns with numerical data types using NumPy's number type
num_cols = data.select_dtypes(include=np.number).columns.tolist()

# Print categorical columns
print("Categorical Variables:")
print(cat_cols)

# Print numerical columns
print("Numerical Variables:")
print(num_cols)
```

```python
# Loop through each numerical column in num_cols
for col in num_cols:
    # Print column name
    print(col)
    
    # Compute and print the skewness of the column (rounded to 2 decimal places)
    print('Skew :', round(data[col].skew(), 2))
    
    # Create a new figure with a specified size
    plt.figure(figsize = (15, 4))
    
    # First subplot: Histogram to visualize the distribution of the numerical column
    plt.subplot(1, 2, 1)    # (1 row, 2 columns, 1st subplot)
    data[col].hist(grid=False)   # Plot histogram without grid
    plt.ylabel('count')   # Label the y-axis
    
    # Second subplot: Boxplot to detect outliers and spread of data
    plt.subplot(1, 2, 2)  # (1 row, 2 columns, 2nd subplot)
    sns.boxplot(x=data[col])  # Seaborn boxplot
    
    #to display the plots
    plt.show()
```

```python
# Create a figure with a grid of 3 rows and 2 columns for subplots
fig, axes = plt.subplots(3, 2, figsize = (18, 18))

# Set a title for the entire figure
fig.suptitle('Bar plot for all categorical variables in the dataset')

# Ploting 6 bar plots (countplots) for categorical variables
# 1st subplot (Top-left): Fuel_Type distribution
sns.countplot(ax = axes[0, 0], x = 'Fuel_Type', data = data, color = 'blue', 
              order = data['Fuel_Type'].value_counts().index);

# 2nd subplot (Top-right): Transmission distribution
sns.countplot(ax = axes[0, 1], x = 'Transmission', data = data, color = 'blue', 
              order = data['Transmission'].value_counts().index);

# 3rd subplot (Middle-left): Owner_Type distribution
sns.countplot(ax = axes[1, 0], x = 'Owner_Type', data = data, color = 'blue', 
              order = data['Owner_Type'].value_counts().index);

# 4th subplot (Middle-right): Location distribution
sns.countplot(ax = axes[1, 1], x = 'Location', data = data, color = 'blue', 
              order = data['Location'].value_counts().index);

# 5th subplot (Bottom-left): Brand distribution (only top 20 brands)
sns.countplot(ax = axes[2, 0], x = 'Brand', data = data, color = 'blue', 
              order = data['Brand'].head(20).value_counts().index);

# 6th subplot (Bottom-right): Model distribution (only top 20 models)
sns.countplot(ax = axes[2, 1], x = 'Model', data = data, color = 'blue', 
              order = data['Model'].head(20).value_counts().index);

# Rotate x-axis labels for better readability
axes[1][1].tick_params(labelrotation=45);
axes[2][0].tick_params(labelrotation=90);
axes[2][1].tick_params(labelrotation=90);
```

```python
# Function for log transformation of the column
def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()
log_transform(data,['Kilometers_Driven','Price'])
#Log transformation of the feature 'Kilometers_Driven'
sns.distplot(data["Kilometers_Driven_log"], axlabel="Kilometers_Driven_log");
```

```python
# Create a new figure with specified dimensions (13x17 inches)
plt.figure(figsize=(13,17))

# Create a pairplot of the data, excluding 'Kilometers_Driven' and 'Price' columns
sns.pairplot(data=data.drop(['Kilometers_Driven','Price'],axis=1))

# Display the pairplot
plt.show()
```
